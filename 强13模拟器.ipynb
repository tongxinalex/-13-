```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create a function to calculate the money needed to get a 13-weapon
def shangshisan(money):

    flag13 = False
    flag12 = False

    while flag13 == False:

        while flag12 == False:
            random_num = float(np.random.rand(1))
            if random_num>0.3:
                money -= 41.38
            else:
                flag12 = True

        if flag12 == True and flag13 == False:
            random_num = float(np.random.rand(1))
            if random_num>0.15:
                flag12 = False
                money -= 4.93
            else:
                flag13 = True

    return money
```

```python
fig, axs = plt.subplots(10,10,figsize=[20,20])

earning = []
earning_per_person = []

for i in range(10):
    for j in range(10):
        
        
        budget = 5000

        customer_num = 20

        money_change = [budget]

        for n in range(customer_num):
            budget = budget + shangshisan(966)
            money_change.append(budget)

        axs[j,i].plot(money_change)
        axs[j,i].plot(money_change, '.', linewidth=0.05)
        axs[j,i].axhline(y = 5000, linestyle='--', color='grey')
        
        if budget<5000:
            axs[j,i].set_facecolor("yellow")
        
#         print(budget-5000, '/', (budget-5000)/customer_num)
        
        earning.append(budget-5000)
        earning_per_person.append((budget-5000)/customer_num)
    
plt.show()
```

```python
fig, axs = plt.subplots(figsize=[2,2])
sns.swarmplot(earning)
axs.set_title('Total earning')
plt.show()

fig, axs = plt.subplots(figsize=[2,2])
sns.swarmplot(earning_per_person)
axs.set_title('Earning per customer')
plt.show()
```
